```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_cleaned.csv')
```

```python
# Drop the day month and year
df.drop(['day','month','year'],axis=1, inplace=True)
```

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df['Classes'].value_counts()
```

```python
# Independent and dependent features
X = df.drop('FWI',axis=1)
y= df['FWI']
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
# Feature selection based on the correlation
X_train.corr()
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(X_train.corr(), annot=True)
plt.show()
```

#### Creating the threshold for the independent data for model selection

```python
def correlation(dataset, threshold):
    col_corr = set()
    col_matrix = dataset.corr()
    for i in range(len(col_matrix.columns)):
        for j in range(i):
            if abs(col_matrix.iloc[i,j] > threshold):
                colname = col_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
# threshold - Domain Expertise
corr_features = correlation(X_train,0.85)
```

```python
corr_features
```

```python
# Drop features when correlation is more than 0.85
X_train.drop(corr_features, axis=1, inplace=True)
X_test.drop(corr_features, axis=1, inplace=True)
X_train.shape , X_test.shape
```

### Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

### Box Plots To understand Effect of Standard Scaler

```python
plt.subplots(figsize=(20,8))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("X Train before Scaling")
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title("X_train after scaling")
```

### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, root_mean_squared_error
from sklearn.metrics import r2_score

regressor = LinearRegression()
regressor.fit(X_train_scaled, y_train)
y_pred = regressor.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("Mean absolute error", mae)
print("Root mean squared error", rmse)
print("R2 score",score)
plt.scatter(y_test, y_pred)
```

### Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error, root_mean_squared_error
from sklearn.metrics import r2_score
lasso = Lasso()
lasso.fit(X_train_scaled, y_train)
y_pred = lasso.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("Mean absolute Error", mae)
print("R2 score ", score)
plt.scatter(y_test, y_pred)
```

### Cross Validation Lasso

```python
from sklearn.linear_model import LassoCV
lassocv =LassoCV(cv=5)
lassocv.fit(X_train_scaled, y_train)
```

```python
lassocv.alpha_
```

```python
y_pred = lassocv.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("Mean absolute error", mae)
print("R2 score", score)
plt.scatter(y_test, y_pred)
```

### Ridge Regression Model

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(X_train_scaled, y_train)
y_pred = ridge.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("Mean absolute Error", mae)
print("R2 score", score)
plt.scatter(y_test, y_pred)
```

```python
from sklearn.linear_model import RidgeCV
ridegcv = RidgeCV(cv=5)
ridegcv.fit(X_train_scaled, y_train)
y_pred = ridegcv.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test , y_pred)
score = r2_score(y_test, y_pred)
print("Mean absolute error", mae)
print("R2 score", score)
plt.scatter(y_test, y_pred)
```
